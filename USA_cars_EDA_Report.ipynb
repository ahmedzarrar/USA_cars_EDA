Importing pandas and numpy in order to perform EDA on the CSV dataset

```python
import pandas as pd
import numpy as np
```

Reading csv file and storing it into a variable named df

```python
df = pd.read_csv("USA_cars_datasets.csv", index_col = False)
df
```

Getting information related to dataset

Dropping any duplicates values from the dataset

```python
df.drop_duplicates(inplace = True)
```

Dropping na values form the datset

```python
df.dropna()
```

As the price can't be less than zero so we are removing those  rows form our dataset

```python
df = df[df['price'] > 0]
```

The mileage column is being filtered to remove values above the 99th percentile, effectively eliminating the top 1% of extreme values (outliers) from the dataset. This is done by keeping only rows where mileage is less than the 99th percentile value.

```python
df = df[df['mileage']<df['mileage'].quantile(0.99)]
```

Importing Libraries for visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

The chart highlights major car brands while grouping less common ones into "Others," simplifying visualization and improving readability.

```python
    brand_counts = df['brand'].value_counts()
    threshold = (5 / 100) * brand_counts.sum()
    major_brands = brand_counts[brand_counts >= threshold]
    other_brands = brand_counts[brand_counts < threshold]
    other_total = other_brands.sum()
    brand_counts_filtered = pd.concat([major_brands, pd.Series({'Others': other_total})])
    plt.figure(figsize=(8, 8))
    wedges, texts, autotexts = plt.pie(brand_counts_filtered, labels=brand_counts_filtered.index, autopct='%1.1f%%', startangle=90)
    if 'Others' in brand_counts_filtered.index:
        other_brands_list = '\n'.join(other_brands.index)
        plt.text(1.2, 0, f'Included in Others:\n{other_brands_list}', fontsize=10, bbox=dict(facecolor='white', alpha=0.6))
    plt.title("Brand Distribution of Cars")
    plt.show()
```

The histogram reveals the distribution of car prices, identifying common price ranges and potential skewness in the dataset.

```python
df['price'].hist(bins=30, edgecolor='black')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

The boxplot visualizes the spread of car prices, highlighting the median, interquartile range, and potential outliers.

```python
sns.boxplot(x=df['price'])
plt.title("Price Boxplot")
plt.show()
```

The bar chart shows the top 10 most common car brands in the dataset, helping identify market dominance and brand popularity.

```python
df['brand'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Car Brands")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

The scatter plot reveals the relationship between mileage and price, typically showing that higher mileage cars tend to have lower prices.

```python
plt.scatter(df['mileage'], df['price'], alpha=0.5)
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.title("Price vs Mileage")
plt.show()
```

The boxplot compares price distributions across different car brands, highlighting price variability, medians, and outliers for each brand.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='brand', y='price', data=df)
plt.xticks(rotation=90)
plt.title("Price Distribution by Brand")
plt.show()
```

The violin plot shows the distribution and density of car prices for each brand, highlighting price spread and concentration areas.

```python
sns.violinplot(x='brand', y='price', data=df)
plt.xticks(rotation=90)
plt.title("Price Variability by Brand")
plt.show()
```

The pair plot visualizes relationships between price, mileage, and year, helping identify trends, correlations, and potential patterns in the dataset.

```python
sns.pairplot(df[['price', 'mileage', 'year']])
plt.show()
```

The heatmap displays correlation strengths between price, mileage, and year, revealing positive or negative relationships that impact car pricing.

```python
sns.heatmap(df[['price', 'mileage', 'year']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

The count plot shows the distribution of title statuses, helping identify the proportion of clean, salvage, rebuilt, and other titles in the dataset.

```python
sns.countplot(x='title_status', hue='title_status', data=df, palette='viridis', legend=False)
plt.title("Distribution of Title Status")
plt.xticks(rotation=45)
plt.show()
```

The stacked bar chart shows the distribution of car brands across different states, highlighting regional preferences and availability of specific brands.

```python
pd.crosstab(df['state'], df['brand']).plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Car Brands Distribution Across States")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Brand", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

The line plot shows the trend of average car prices over the years, helping identify whether prices increase or decrease with newer models.

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='year', y='price', data=df, estimator='mean', errorbar=None, marker='o')
plt.title("Average Car Price Over Years")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

The bar chart highlights the top 10 most expensive car brands based on average price, revealing premium brands in the dataset.

```python
df.groupby("brand")["price"].mean().nlargest(10).plot(kind='bar', color='goldenrod')
plt.title("Top 10 Most Expensive Car Brands")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

The bar chart displays the average car price in each state, revealing regional pricing differences and potential market trends.

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=df.groupby("state")["price"].mean().index, y=df.groupby("state")["price"].mean().values)
plt.xticks(rotation=90)
plt.title("Average Car Price by State")
plt.xlabel("State")
plt.ylabel("Average Price")
plt.show()
```
